```python
# Importing the basic libraries

import pandas as pd
import numpy as np
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine
import warnings
```

```python
books = pd.read_csv("C:/Users/Aboli/Downloads/book.csv")
```

```python
books.head()
```

```python
# Renaming the columns

books_data = books.rename({'User.ID':'userid','Book.Title':'booktitle','Book.Rating':'bookrating'},axis = 1)
```

```python
books_data.head()
```

```python
# Dropping the column

books1 = books_data.drop(['Unnamed: 0'], axis = 1)
```

```python
books1.head()
```

```python
books_data1 = books1.pivot_table(index = 'userid',
                        columns = 'booktitle',
                        values = 'bookrating').reset_index(drop = True)
```

```python
books_data1.head()
```

```python
books_data1.index = books1.userid.unique()
```

```python
# Filling the NaN values with 0

books_data1.fillna(0, inplace = True)
```

```python
books_data1.head()
```

```python
warnings.filterwarnings("ignore")

user = 1 - pairwise_distances(books_data1.values, metric = 'cosine')
```
